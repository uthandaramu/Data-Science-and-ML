### Series

```python
import pandas as pd
data = [1, 2, 3, 4]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
#customized series num
data1 = {
    'a' : 1,
    'b' : 2,
    'c' : 3,
    'd' : 4
}

series1 = pd.Series(data1)
print(series1)
```

```python
#customized series num - 2
data2 = [1, 2, 3, 4]
index = ['a', 'b', 'c', 'd']
series2 = pd.Series(data2, index=index)
print(series2)
```

### Handling Null values

```python
df[df['Units Sold'].isnull()]
```

```python
malawi_country = df[df['Country'] == 'Malawi']
malawi_country.head()
```

```python
malawi_country['Units Sold'].mean()
```

```python
malawi_country.describe()
```

```python
malawi_mean = df[df['Country'] == 'Malawi']['Units Sold'].mean()
```

```python
malawi_mean
```

```python
df['Units Sold NA'] = df['Units Sold']
df.loc[df['Country']=='Malawi', 'Units Sold NA'] = df.loc[df['Country']=='Malawi', 'Units Sold NA'].fillna(malawi_mean)
```

```python
df[df['Units Sold'].isnull()]
```

```python
liberia_mean = df[df['Country'] == 'Liberia']['Units Sold'].mean()
```

```python
liberia_mean
```

```python
df.loc[df['Country']=='Liberia', 'Units Sold NA'] = df.loc[df['Country']=='Liberia', 'Units Sold NA'].fillna(liberia_mean)
```

```python
df[df['Units Sold'].isna()]
```

```python
slovenia_mean = df.loc[df['Country']=='Slovenia', 'Units Sold'].mean()
```

```python
slovenia_mean
```

```python
df.iloc[227, 14] = slovenia_mean
```

```python
df[df['Units Sold'].isna()]
```

```python
df.isnull().sum()
```

```python
df[df['Unit Cost'].isna()]
```

```python
df['Unit Cost'] = df['Unit Cost'].fillna(df['Unit Cost'].mean())
```

```python
df[df['Unit Cost'].isna()]
```

```python
df['Units Sold'] = df['Units Sold NA']
df.drop('Units Sold NA', axis=1)
```

```python
df['Total Revenue'] = df['Total Revenue'].fillna(df['Total Revenue'].mean())
df['Total Cost'] = df['Total Cost'].fillna(df['Total Cost'].mean())
df['Total Profit'] = df['Total Profit'].fillna(df['Total Profit'].mean())
```

```python
df.isnull().sum()
```

```python
#rename
df.rename(columns={'Units_Sold_NA' : 'Units Sold NA', 'Total Profit' : 'Total_Profit'}, inplace=True)
```

```python
#Using lambda expression
df['New Unit Cost'] = df['Unit Cost'].apply(lambda x:x*2)
```

```python
df.head()
```

```python
#Data aggregation and grouping
grouped_mean = df.groupby('Country')['Units Sold'].mean()
```

```python
grouped_mean
```

```python
#grouping multiple columns
grouped_item_mean = df.groupby(['Country', 'Sales Channel'])['Total Revenue'].mean()
```

```python
grouped_item_mean
```

```python
#aggregating multiple columns
grouped_aggregation = df.groupby('Sales Channel')['Units Sold'].agg(['mean', 'sum', 'count'])
grouped_aggregation
```

```python
#Merging and joining dataframes
df1 = pd.DataFrame({
    'key':['a', 'b', 'c'],
    'value':[1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['c', 'd', 'e'],
    'value': [4, 5, 6]
})

pd.merge(df1, df2, on='key', how='right')
```
