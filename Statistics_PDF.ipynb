### Probability Distribution

```python
#normal/Gaussian
mu = 0.5
sigma = 0.2
normal_distributed_data = np.random.normal(mu, sigma, 1000)
```

```python
sns.histplot(normal_distributed_data, kde=True)
```

```python
#Log-normal
mu = 3
sigma = 1.0

lognormal_distributed_data = np.random.lognormal(mu, sigma, 500)

sns.histplot(lognormal_distributed_data, kde=True)
```

```python
normal_from_log = np.log(lognormal_distributed_data)
```

```python
sns.histplot(normal_from_log, kde=True)
```

```python
#Uniform Distribution
uniform_data = np.random.uniform(size=10000)
```

```python
sns.histplot(uniform_data, kde=True)
```

```python
#bernouli data
p =0.3
sample_size = 1000
bernouli_data = np.random.choice([0, 1], size = sample_size, p=[1-p, p])
sns.histplot(bernouli_data, bins=2)
```

```python
#poisson
lamda = 12
sample_size = 1000
poison_data = np.random.poisson(lamda, sample_size)
sns.histplot(poison_data)
```

```python
#Binomial distribution
binomial_data = np.random.binomial(n=100, p=0.5, size=1000)
sns.histplot(binomial_data)
```
