```python
import matplotlib.pyplot as plt
```

### Basic Line Plot

```python
x = [1,2,3,4,5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y, marker='s', linestyle='--', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### Multiple plots

```python
x = [1, 2, 3, 4, 5, 6]
y = [7, 8, 9, 10, 11, 12]
z = [13, 14, 15, 16, 17, 18]

plt.figure(figsize=(6, 5))

plt.subplot(2,2,1)
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(2,2,2)
plt.plot(y,z)
plt.xlabel('Y')
plt.ylabel('Z')

plt.subplot(2,2,3)
plt.plot(z, x)
plt.xlabel('Z')
plt.ylabel('X')

plt.subplot(2,2,4)
plt.plot(x, z)
plt.xlabel('X')
plt.ylabel('Z')
```

### Bar Plot

```python
cat = ['a', 'b', 'c', 'd']
val = [5, 2, 13, 1]

plt.bar(cat, val, color = 'teal')
plt.show()
```

### Histogram

```python
data = [1, 2, 2, 3, 3, 3, 4, 5, 6, 7, 8, 11, 11, 11, 2, 5, 2, 1, 9, 12]
plt.hist(data, edgecolor='black', color='pink', bins=6)
plt.show()
```

### Scatterplot

```python
val = [2, 4, 1, 9, 3, 5]
val2 = [1, 3, 5, 12, 2, 4]
key = [1, 2, 3, 4, 5, 6]
plt.scatter(key, val, marker='x', color = 'red')
plt.show()
```

### Pie Chart

```python
cat = ['a', 'b', 'c', 'd', 'e']
val = [15, 2, 13, 11, 20]
colors = ['yellowgreen', 'lightcoral', 'skyblue', 'teal', 'gold']
explode = (0.1, 0, 0, 0, 0)
plt.pie(val, labels=cat, explode=explode, colors=colors, autopct='%.2f%%', shadow=True)
```
