```python
#Decorators
def decor(funct):
    def inner(name):
        if name == "Jesse":
            print ("Good Morning Jesse")
        else:
            funct(name)
    return inner

@decor
def wish(name):
    print("Bad Morning", name)
```

```python
wish("Jesse")
```

```python
wish("Pinkman")
```

```python
#Decorators
def decor(funct):
    def inner(name):
        if name == "Jesse":
            print ("Good Morning Jesse")
        else:
            funct(name)
    return inner

def wish(name):
    print("Bad Morning", name)

decor_arr = decor(wish)
```

```python
wish("Jesse")
```

```python
decor_arr("Jesse")
```

```python
#Decorator chaining

def decor1(func):
    def double():
        x = func()
        return x * 2
    return double

def decor2(func):
    def square():
        x = func()
        return x ** 2
    return square

@decor2
@decor1
def double_square():
    return 10
```

```python
double_square()
```

```python
#Decorator chaining

def decor1(func):
    def double(num):
        x = func(num)
        return x * 2
    return double

def decor2(func):
    def square(num1):
        x = func(num1)
        return x ** 2
    return square

@decor2
@decor1
def double_square(n):
    return n
```

```python
double_square(3)
```

```python
double_square(21)
```

```python
#Generators
def mygen():
    yield "A"
    yield "B"
    yield "C"

g = mygen()
```

```python
def firstn(num):
    n = 1
    while n <= num:
        yield n
        n += 1
values = firstn(10)

for i in values:
    print(i)
```

```python
values2 = firstn(20)
l = list(values2)
print(l)
```
